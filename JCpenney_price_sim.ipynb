# JCPenney Price Simulator

## Project Overview

This notebook builds a simple price simulator using JCPenney product and pricing data.

The goal is to answer:

> If I change the price of a product, what happens to estimated sales?

The simulator takes 100 products and compares two pricing scenarios:

- **Scenario A:** 20% price cut for a flash sale
- **Scenario B:** 10% price increase going into a new season

Because the dataset does not include actual units sold, this simulator estimates sales movement using review-count-based baseline demand and category-level price elasticity assumptions.

```python


# -----------------------------
# USER INPUTS
# -----------------------------

zip_path = "data/sample_data (2).zip"

NUM_PRODUCTS = 100

SCENARIO_A_PRICE_CHANGE = -0.20
SCENARIO_B_PRICE_CHANGE = 0.10

SCENARIO_A_NAME = "Scenario A: 20% Flash Sale Discount"
SCENARIO_B_NAME = "Scenario B: 10% Seasonal Price Increase"
```

## 1. Load and Combine the Data

The raw data is split across many CSV files inside the ZIP folder. Each file represents a category/date scrape. I combine all CSV files into one dataset and preserve the scrape date for later cleaning.

```python
all_dfs = []

with zipfile.ZipFile(zip_path, "r") as z:
    csv_files = [f for f in z.namelist() if f.endswith(".csv")]

    for file in csv_files:
        with z.open(file) as f:
            temp = pd.read_csv(f)

            date_match = re.search(r"data/(\d{4}-\d{2}-\d{2})/", file)
            temp["scrape_date"] = date_match.group(1) if date_match else None

            temp["source_file"] = file

            all_dfs.append(temp)

df = pd.concat(all_dfs, ignore_index=True)

print("Raw combined dataset shape:", df.shape)
df.head()
```
